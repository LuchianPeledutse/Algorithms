### Merge and insetion sort time comparison

```python
#Merge that is needed for MergeSort
def Merge(A,p,q,r):
	#len(A) > 1 as p<=q<r
	n1 = q-p+1
	n2 = r-q
	#create left and right sequences
	L = [1 for _ in range(n1)]
	R = [1 for _ in range(n2)]
	#filling them up
	for k in range(n1):
		L[k] = A[p+k]
	for k in range(n2):
		R[k] = A[q+k+1]
	#sorting in non-decreasing order
	i,j = 0,0
	for ind in range(p,r+1):
		if i+1 > n1:
			A[ind] = R[j]
			j += 1
		elif j+1 > n2:
			A[ind] = L[i]
			i += 1
		else:
			if L[i] <= R[j]:
				A[ind] = L[i]
				i += 1
			else:
				A[ind] = R[j]
				j += 1

#merge_sort
def MergeSort(A,p,r):
	#here we have p<=r
	#if p = r nothing is done
	if p<r:
		q = int((p+r)/2)
		#q satisfies p<=q<r
		MergeSort(A,p,q)
		MergeSort(A,q+1,r)
		Merge(A,p,q,r)

#insertion_sort
def insertion_sort(A):
	N = len(A)
	for j in range(1,N):
		key = A[j]
		i = j-1
		while i>=0 and A[i]>key:
			A[i+1] = A[i]
			i -= 1
		A[i+1] = key
```

```python
def test_merge(N):
	for k in range(N):
		rand_ar = [rd.randint(1,99) for _ in range(11)]
		py_sort = list(sorted(rand_ar))
		print(f'Python_sort: {py_sort}')
		print(f'Initital_list: {rand_ar}')
		MergeSort(rand_ar,0,len(rand_ar)-1)
		print(f'MergeSort: {rand_ar}',end = '\n\n')
		if rand_ar != list(sorted(rand_ar)):
			print('#################WRONG#################')


def test_ins_sort(N):
	for k in range(N):
		rand_ar = [rd.randint(1,99) for _ in range(10)]
		py_sort = list(sorted(rand_ar))
		print(f'Python_sort: {py_sort}')
		print(f'Initital_list: {rand_ar}')
		insertion_sort(rand_ar)
		print(f'Insertion_sort: {rand_ar}',end = '\n\n')
		if rand_ar != list(sorted(rand_ar)):
			print('#################WRONG#################')
```

```python
N = 5
test_merge(N)
test_ins_sort(N)
```

```python
#time of 100_000 elements of MergeSort
start_time = time.time()
some_ar = [rd.randint(1,99) for _ in range(100_000)]
MergeSort(some_ar,0,len(some_ar)-1)
end_time = time.time()
print(f'Mergesort_time:{end_time-start_time:.6f} seconds')
```

```python
#time of 100_000 elements of InsertionSort
start_time = time.time()
some_ar = [rd.randint(1,99) for _ in range(100_000)]
insertion_sort(some_ar)
end_time = time.time()
print(f'Insertion_sort_time:{end_time-start_time:.6f} seconds')
```

### Two sum n^2 and nlogn comparisons

```python
def measure_execution_time(func, args_list):
    start_time = time.perf_counter()  
    func(*args_list) 
    end_time = time.perf_counter()
    return round(end_time - start_time,5)
```

```python
def time_test(N = 10):
    for _ in range(N):
        rand_shape = rd.randint(10_000,25_000)
        main_list = [rd.randint(1000,1500) for  _ in range(rand_shape)]
        random_find = rd.randint(1000,1500)
        print(f'Time of O(n^2): {measure_execution_time(n2_2_sum,[main_list,random_find])}')
        print(f'Time of O(nlogn): {measure_execution_time(nlogn_2_sum,[main_list,random_find])}',end = '\n\n')
```

```python
time_test(N = 5)
```

```python
def match_test(N = 100):
    for _ in range(N):
        rand_shape = rd.randint(10_000,25_000)
        main_list = [rd.randint(500,1500) for  _ in range(rand_shape)]
        random_find = rd.randint(1000,1500)
        print(f'Test # {_+1}')
        res1 = n2_2_sum(main_list,random_find)
        res2 = nlogn_2_sum(main_list,random_find)
        if  res1 == res2:
            print('PASSED')
            print(res1,res2)
        else:
            print('###WRONG###')
            break
```
